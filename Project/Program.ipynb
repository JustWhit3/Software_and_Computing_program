## Splitting of the data into training, test and validation set

```python
train_data = data_full[data_full['KaggleSet']=='t']
validation_data = data_full[data_full['KaggleSet']=='b']
test_data =  data_full[data_full['KaggleSet']=='v']

#I have to construct binary arrays for my network.
y_train = pd.get_dummies(train_data['Label']).values.astype(float) #signal = [0,1], background = [1,0]
y_val = pd.get_dummies(validation_data['Label']).values.astype(float)
y_test = pd.get_dummies(test_data['Label']).values.astype(float)

##### for BDT #### It needs 1D classes and can handle strings, so we can use the 's' or 'b' labels
y_train_BDT = train_data['Label']
y_val_BDT = validation_data['Label']
y_test_BDT = test_data['Label']
##################

#Let's test how many data I've for each subset.
del(data_full)  
len(train_data), len(validation_data), len(test_data)
```

```python
X = train_data.iloc[:,1:-4]
X_val = validation_data.iloc[:,1:-4]
X_test = test_data.iloc[:,1:-4]

weights_val = validation_data['KaggleWeight']
weights_test = test_data['KaggleWeight']

Event_IDs_val = validation_data['EventId']
Event_IDs_test = test_data['EventId']

del train_data, validation_data, test_data
```

## Feature Engineering
The problem is that the "phi" variables have a signal distribution that is very similar to the background one. So it's better to consider their linear combination (difference in this case) to make them useful in my classification.
### Replace Phi variables by their difference
+ `Delta_phi_tau_lep` = `PRI_tau_phi` - `PRI_lep_phi` (not helpful for 2 jets category)
+ `Delta_phi_met_lep` = `PRI_met_phi` - `PRI_lep_phi`
+ `Delta_phi_jet_jet` = `PRI_jet_leading_phi` - `PRI_jet_subleading_phi`
    + Drop `PRI_tau_phi`,  `PRI_lep_phi`, `PRI_met_phi`, `PRI_jet_leading_phi` and `PRI_jet_subleading_phi`


## Add Delta eta tau lep

```python
X['Delta_phi_tau_lep'] = abs(X['PRI_tau_phi']-X['PRI_lep_phi'])
X_val['Delta_phi_tau_lep'] = abs(X_val['PRI_tau_phi']-X_val['PRI_lep_phi'])
X_test['Delta_phi_tau_lep'] = abs(X_test['PRI_tau_phi']-X_test['PRI_lep_phi'])


X['Delta_phi_jet_jet'] = abs(X['PRI_jet_leading_phi']-X['PRI_jet_subleading_phi'])
X_val['Delta_phi_jet_jet'] = abs(X_val['PRI_jet_leading_phi']-X_val['PRI_jet_subleading_phi'])
X_test['Delta_phi_jet_jet'] = abs(X_test['PRI_jet_leading_phi']-X_test['PRI_jet_subleading_phi'])


X['Delta_phi_met_lep'] = abs(X['PRI_met_phi']-X['PRI_lep_phi'])
X_val['Delta_phi_met_lep'] = abs(X_val['PRI_met_phi']-X_val['PRI_lep_phi'])
X_test['Delta_phi_met_lep'] = abs(X_test['PRI_met_phi']-X_test['PRI_lep_phi'])

X['Delta_eta_tau_lep'] = abs(X['PRI_tau_eta']-X['PRI_lep_eta'])
X_val['Delta_eta_tau_lep'] = abs(X_val['PRI_tau_eta']-X_val['PRI_lep_eta'])
X_test['Delta_eta_tau_lep'] = abs(X_test['PRI_tau_eta']-X_test['PRI_lep_eta'])
```

## I) BDT Classification
I first will use a BDT model without splitting according to jet number. The class used for this is: `HistGradientBoostingClassifier` from `Sklearn`.

```python
from sklearn.preprocessing import StandardScaler

#Let's first scale my data.
standard = StandardScaler()
standard.fit(X)
X_standard = standard.transform(X)
X_val_standard = standard.transform(X_val)
X_test_standard = standard.transform(X_test)


from sklearn.experimental import enable_hist_gradient_boosting 
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
BDT = HistGradientBoostingClassifier(max_iter=90, verbose=1, l2_regularization=0.5, learning_rate=.1, max_leaf_nodes=50, 
                                     random_state=45, max_depth=15, max_bins=50)
BDT.fit(X_standard, y_train_BDT)

y_pred_val = BDT.predict_proba(X_val_standard)
y_pred_test = BDT.predict_proba(X_test_standard)

plot_AMS(y_pred_val, y_val, weights_val)
plt.title("Validation Set")
plt.show()
plot_AMS(y_pred_test, y_test, weights_test)
plt.title("Test Set")
plt.show()

plot_distributions(y_pred_val, y_val, 50, weighted=False, weights = None)

del X_standard, X_val_standard, X_test_standard
```

I will split the results just to be able to combine them with the DNN result later

```python
BDT_0jets_val = y_pred_val[ X_val['PRI_jet_num']==0 ]
BDT_1jet_val = y_pred_val[ X_val['PRI_jet_num']==1 ]
BDT_2jets_val = y_pred_val[ X_val['PRI_jet_num']>=2 ]

y_pred_BDT_val = np.concatenate((BDT_0jets_val, BDT_1jet_val, BDT_2jets_val))


BDT_0jets_test = y_pred_test[ X_test['PRI_jet_num']==0 ]
BDT_1jet_test = y_pred_test[ X_test['PRI_jet_num']==1 ]
BDT_2jets_test = y_pred_test[ X_test['PRI_jet_num']>=2 ]

y_pred_BDT_test = np.concatenate((BDT_0jets_test, BDT_1jet_test, BDT_2jets_test))
```

## II) DNN Classification
### Build subsets according to the number of jets because of the different event topologies

Since the BDT model seems to work better without dropping the features and the DNN seems to work better after dropping, I put this here. 

Moreover, the HistBDT can handle these -999 values very well, but for the DNN it helped to set them to the mean.

Now I have to construct three subsets according to the number of jets:
+ 0 jets.
+ 1 jet.
+  \>= 2 jets.

I drop the features in which I have a value equal to -999.

```python
X['DER_mass_MMC'][X['DER_mass_MMC']==-999.] = np.mean(X['DER_mass_MMC'][X['DER_mass_MMC']!=-999.])
X_val['DER_mass_MMC'][X_val['DER_mass_MMC']==-999.] = np.mean(X_val['DER_mass_MMC'][X_val['DER_mass_MMC']!=-999.])
X_test['DER_mass_MMC'][X_test['DER_mass_MMC']==-999.] = np.mean(X_test['DER_mass_MMC'][X_test['DER_mass_MMC']!=-999.])
#setting the -999 value of MMC to the mean improved classification

X = X.drop(['PRI_tau_phi', 'PRI_lep_phi', 'PRI_jet_leading_phi', 'PRI_jet_subleading_phi'
            #,'PRI_met_phi'
            #,'PRI_tau_eta', 'PRI_lep_eta'
            ####### dropping these makes it worse
           ], axis=1)
X_val = X_val.drop(['PRI_tau_phi', 'PRI_lep_phi', 'PRI_jet_leading_phi', 'PRI_jet_subleading_phi'
                    #, 'PRI_met_phi'
                    #,'PRI_tau_eta', 'PRI_lep_eta'
                    ####### dropping these makes it worse
                  ], axis=1)
X_test = X_test.drop(['PRI_tau_phi', 'PRI_lep_phi', 'PRI_jet_leading_phi', 'PRI_jet_subleading_phi'
                    #, 'PRI_met_phi'
                    #,'PRI_tau_eta', 'PRI_lep_eta'
                    ####### dropping these makes it worse
                  ], axis=1)
```

```python
##### 0 jets
X_0jets = X[ X['PRI_jet_num']==0 ].drop(['DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 
                                         'DER_lep_eta_centrality', 'PRI_jet_num', 'PRI_jet_leading_pt', 
                                         'PRI_jet_leading_eta', 'PRI_jet_subleading_pt', 
                                         'PRI_jet_subleading_eta', 'PRI_jet_all_pt', 'PRI_jet_num',
                                         'Delta_phi_jet_jet', 'PRI_jet_leading_phi', 'PRI_jet_subleading_phi'], 
                                        axis =1, errors='ignore')
X_val_0jets = X_val[ X_val['PRI_jet_num']==0 ].drop(['DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 
                                         'DER_lep_eta_centrality', 'PRI_jet_num', 'PRI_jet_leading_pt', 
                                         'PRI_jet_leading_eta', 'PRI_jet_subleading_pt', 
                                         'PRI_jet_subleading_eta', 'PRI_jet_all_pt', 'PRI_jet_num',
                                         'Delta_phi_jet_jet', 'PRI_jet_leading_phi', 'PRI_jet_subleading_phi'], 
                                        axis =1, errors='ignore')
X_test_0jets = X_test[ X_test['PRI_jet_num']==0 ].drop(['DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 
                                         'DER_lep_eta_centrality', 'PRI_jet_num', 'PRI_jet_leading_pt', 
                                         'PRI_jet_leading_eta', 'PRI_jet_subleading_pt', 
                                         'PRI_jet_subleading_eta', 'PRI_jet_all_pt', 'PRI_jet_num',
                                         'Delta_phi_jet_jet', 'PRI_jet_leading_phi', 'PRI_jet_subleading_phi'], 
                                        axis =1, errors='ignore')

y_train_0jets = y_train[ X['PRI_jet_num']==0 ]
y_val_0jets = y_val[ X_val['PRI_jet_num']==0 ]
y_test_0jets = y_test[ X_test['PRI_jet_num']==0 ]
weights_0jets_val = weights_val[ X_val['PRI_jet_num']==0 ].reset_index(drop=True)
weights_0jets_test = weights_test[ X_test['PRI_jet_num']==0 ].reset_index(drop=True)


##### 1 jet
X_1jet = X[ X['PRI_jet_num']==1 ].drop(['DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 
                                        'PRI_jet_subleading_pt', 'PRI_jet_subleading_eta', 
                                        'DER_lep_eta_centrality', 'PRI_jet_num',
                                        'Delta_phi_jet_jet', 'PRI_jet_subleading_phi'], axis=1, errors='ignore')
X_val_1jet = X_val[ X_val['PRI_jet_num']==1 ].drop(['DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 
                                        'PRI_jet_subleading_pt', 'PRI_jet_subleading_eta', 
                                        'DER_lep_eta_centrality', 'PRI_jet_num',
                                        'Delta_phi_jet_jet', 'PRI_jet_subleading_phi'], axis=1, errors='ignore')
X_test_1jet = X_test[ X_test['PRI_jet_num']==1 ].drop(['DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 
                                        'PRI_jet_subleading_pt', 'PRI_jet_subleading_eta', 
                                        'DER_lep_eta_centrality', 'PRI_jet_num',
                                        'Delta_phi_jet_jet', 'PRI_jet_subleading_phi'], axis=1, errors='ignore')

y_train_1jet = y_train[ X['PRI_jet_num']==1 ]
y_val_1jet = y_val[ X_val['PRI_jet_num']==1 ]
weights_1jet_val = weights_val[ X_val['PRI_jet_num']==1 ].reset_index(drop=True)
y_test_1jet = y_test[ X_test['PRI_jet_num']==1 ]
weights_1jet_test = weights_test[ X_test['PRI_jet_num']==1 ].reset_index(drop=True)

#y_train_BDT_1jet = y_train_BDT[ X['PRI_jet_num']==1 ]
#y_val_BDT_1jet = y_val_BDT[ X_val['PRI_jet_num']==1 ]


##### 2 and 3 jets
X_2jets = X[ X['PRI_jet_num']>=2 ].drop(['PRI_jet_num',
                                        ], axis=1, errors='ignore')
X_val_2jets = X_val[ X_val['PRI_jet_num']>=2 ].drop(['PRI_jet_num',
                                                     ], axis=1, errors='ignore')
X_test_2jets = X_test[ X_test['PRI_jet_num']>=2 ].drop(['PRI_jet_num',
                                                     ], axis=1, errors='ignore')
y_train_2jets = y_train[ X['PRI_jet_num']>=2 ]
y_val_2jets = y_val[ X_val['PRI_jet_num']>=2 ]
y_test_2jets = y_test[ X_test['PRI_jet_num']>=2 ]
weights_2jets_val = weights_val[ X_val['PRI_jet_num']>=2 ].reset_index(drop=True)
weights_2jets_test = weights_test[ X_test['PRI_jet_num']>=2 ].reset_index(drop=True)

#y_train_BDT_2jets = y_train_BDT[ X['PRI_jet_num']>=2  ]
#y_val_BDT_2jets = y_val_BDT[ X_val['PRI_jet_num']>=2  ]

del X, X_val, X_test
```

## 2-jets Classification

I first started scaling my data.

```python
standard_2jets = StandardScaler()
standard_2jets.fit(X_2jets)
X_2jets_standard = standard_2jets.transform(X_2jets)
X_val_2jets_standard = standard_2jets.transform(X_val_2jets)
X_test_2jets_standard = standard_2jets.transform(X_test_2jets)
```

Now let's perform a DNN

```python
np.random.seed(42)
DNN_2jets = make_model([64, 128, 64, 64, 32, 8], 'relu', 0.1, 'Adam', 'L2', 0.0001, X_2jets.shape[-1])

early_stopping = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=10, verbose=0, mode='auto', baseline=None, restore_best_weights=True)

history = DNN_2jets.fit(X_2jets_standard, y_train_2jets, batch_size=256, epochs=50, verbose=1, 
                  validation_data=(X_val_2jets_standard, y_val_2jets), callbacks = [early_stopping], class_weight = None)

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='lower right')
plt.show()

y_pred_2jets_val = DNN_2jets.predict(X_val_2jets_standard)
y_pred_2jets_test = DNN_2jets.predict(X_test_2jets_standard)
plot_AMS(y_pred_2jets_val, y_val_2jets, weights_2jets_val)
plt.show()
plt.grid()
plot_AMS(y_pred_2jets_test, y_test_2jets, weights_2jets_test)
plt.grid()
```
